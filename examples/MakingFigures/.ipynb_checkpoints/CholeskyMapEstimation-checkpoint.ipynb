```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import scipy
```

```python
with open('../Gaussian/CholeskyMap_learning.pkl', 'rb') as handle:
    all_info = pickle.load(handle)
print(all_info.keys())
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
plt.rc('text', usetex=True)
plt.rc('font', size=12)
fontsize=16
methods = ['NN','OTT'] 
method_label = [r'$\widehat{T}_{\mathrm{NN},t}$',r'$\widehat{T}_{\varepsilon,t}$']
linestyles = ['-','-']
colors = ['#ff7f0e','#d62728']
plt.figure()
for (m,method) in enumerate(methods):
    L2_err_method = all_info['L2_dict'][method]
    L2_err_mean   = np.mean(L2_err_method,axis=0)
    L2_err_std    = np.std(L2_err_method,axis=0)
    # plot result
    plt.errorbar(all_info['n_samples'], L2_err_mean, yerr=L2_err_std,label=str(method_label[m]),\
                    color=colors[m],linestyle='-')
plt.xscale('log')
plt.yscale('log')
plt.ylim([1e-2,4e-1])
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.xlabel(r'$\mathrm{Number}\;\mathrm{of}\;\mathrm{samples}$ $n$',fontsize=fontsize)
plt.ylabel(r'$\|\widehat{T}_{t} - T_{CB}\|_{L^2(\rho)}^2$',fontsize=fontsize)
plt.legend(fontsize=17)
plt.show()
#plt.savefig('Gaussian_convergence.pdf',dpi=300,bbox_inches='tight')
```
