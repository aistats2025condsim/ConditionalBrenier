```python
%load_ext autoreload
%autoreload 2

#from pykeops_utils import *
import torch
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})

import scipy.io as sio
import scipy.stats as stats

#from matplotlib import cm
```

```python
# load generate samples data 
gen_samples = np.load('samples_eps005_beta005_tol_1eminus4.npy')
print(gen_samples.shape)

# load target samples
target_samples = torch.load('targetdatascaled_eps005_beta005_tol_1eminus4.pt', map_location=torch.device('cpu')).numpy()
print(target_samples.shape)

# load MCMC samples
mcmc_samples = np.load('../../lotka_volterra/mcmc_samps.npz')['x_samps']
print(mcmc_samples.shape)

# load true sample
data = sio.loadmat('../../lotka_volterra/problem_data.mat')
xtrue = data['xtrue'][0,:]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
plt.rc('text', usetex=True)
plt.rc('font', size=12)
symbols = [r'$\alpha$',r'$\beta$',r'$\gamma$',r'$\delta$']
limits = [[0.5,1.3],[0.02,0.07],[0.7,1.5],[0.025,0.065]]
tick_fontsize = 10

def plot_matrix(x_samps, limits, xtrue=None, symbols=None):
    dim = x_samps.shape[1]
    plt.figure(figsize=(9,9))
    for i in range(dim):
        for j in range(i+1):
            ax = plt.subplot(dim,dim,(i*dim)+j+1)
            if i==j:
                plt.hist(x_samps[:,i], bins=40, density=True)
                if xtrue is not None:
                    plt.axvline(xtrue[i], color='r', linewidth=3)
                plt.xlim(limits[i])
                plt.xticks(fontsize=tick_fontsize)
                plt.yticks(fontsize=tick_fontsize)
            else:
                plt.plot(x_samps[:,j], x_samps[:,i], '.k', markersize=.04, alpha=0.1)
                if xtrue is not None:
                    plt.plot(xtrue[j], xtrue[i], '.r', markersize=8, label='Truth')
                # Peform the kernel density estimate
                xlim = limits[j]
                ylim = limits[i]
                xx, yy = np.mgrid[xlim[0]:xlim[1]:100j, ylim[0]:ylim[1]:100j]
                positions = np.vstack([xx.ravel(), yy.ravel()])
                kernel = st.gaussian_kde(x_samps[:,[j,i]].T)
                f = np.reshape(kernel(positions), xx.shape)
                ax.contourf(xx, yy, f, cmap='Blues')
                plt.ylim(limits[i])
                plt.yticks(fontsize=tick_fontsize)
                #plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
            plt.xlim(limits[j])
            plt.xticks(fontsize=tick_fontsize)
            plt.tick_params(axis='y', which='major', pad=0)
            #plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
            if symbols is not None:
                if j==0:
                    plt.ylabel(symbols[i], size=20)
                if i==len(xtrue)-1:
                    plt.xlabel(symbols[j], size=20)


max_samples = min(25000, gen_samples.shape[0])
plot_matrix(gen_samples[:max_samples,:], limits, xtrue, symbols)
plt.savefig('DLV_EOTposterior.pdf', bbox_inches='tight')
plt.savefig('DLV_EOTposterior.png', bbox_inches='tight')
max_samples = min(25000, mcmc_samples.shape[0])
thin_factor = np.floor((mcmc_samples.shape[0] - 50000)/max_samples)
print(mcmc_samples[50000::int(thin_factor),:].shape)
plot_matrix(mcmc_samples[50000::int(thin_factor),:], limits, xtrue, symbols)
plt.savefig('DLV_MCMCposterior.pdf', bbox_inches='tight')
plt.savefig('DLV_MCMCposterior.png', bbox_inches='tight')
```
