```python
import sys
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

sys.path.insert(1, '../../data/')
from datasets import Banana
sys.path.insert(1, '../../methods/')
from TriangularEstimators import ScaledCostOT

plt.rcParams.update({'font.size': 20})

np.random.seed(69)
```

## Define dataset

```python
# define number of samples
N = 7500

# define class
pi = Banana(reverse=True)

# define source samples
X = pi.sample_joint(N)
X_source = np.hstack((X[:,0][:,None], np.random.randn(N,1)))

# define target samples
X_target = pi.sample_joint(N)
```

## Plot joint

```python
x1_dom = [-5,10]
x2_dom = [-3.5,3.5]

plt.figure(figsize=(3,6))
plt.scatter(X_target[:,1], X_target[:,0], s=4)
plt.hlines(3, x2_dom[0], x2_dom[1], color='k', linewidth=3)
plt.hlines(-0.5, x2_dom[0], x2_dom[1], color='k', linewidth=3)
plt.xlabel('$x_2$')
plt.ylabel('$x_1$')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(x1_dom[0], x1_dom[1])
plt.xlim(x2_dom[0], x2_dom[1])
plt.savefig('banana_target.png', dpi=600, bbox_inches='tight')
```

```python
dim_x = dim_y = 1
beta = 6e-2
```

```python
## ICNN model
n_iters_icnn = 5000 ### number of training iterations
ot_icnn = ScaledCostOT(dx1=dim_x,dx2=dim_y,beta=beta,estimator='ICNN', n_iters=n_iters_icnn) ## move this as a **args
ot_icnn.fit(source=X_source, target=X_target)
print('Done ICNN')

## Nearest-Neighbor estimator
if N <= 10000:
    ot_nn = ScaledCostOT(dx1=dim_x,dx2=dim_y,beta=beta,estimator='NN')
    ot_nn.fit(source=X_source, target=X_target)
    print('Done NN')

## Entropic map estimator
eps_ott = beta/5
n_inters_ott = 5000
ot_ott = ScaledCostOT(dx1=dim_x,dx2=dim_y,beta=beta,estimator='OTT',eps=eps_ott)
ot_ott.fit(source=X_source, target=X_target, max_iter=n_inters_ott)
print('Done OTT')
```

```python
import matplotlib.pyplot as plt
import jax.numpy as jnp
plt.rcParams.update({'font.size': 20})
plt.rc('text', usetex=True)
plt.rc('font', size=12)

load_data = False

Ntest = int(5000)
y_locs = [-0.5,3]
nbins = 40
lw = 3
x_dom = [-3,3]

# Share a Y axis with each row of subplots
plt.figure()
fig, ax = plt.subplots(3,2,figsize = (5,7))

for i,y in enumerate(y_locs):
    data_cond_y = jnp.hstack((jnp.ones((Ntest,1))*y, jnp.array(np.random.randn(Ntest,1))))
    # evaluate conditional density
    xx = np.linspace(x_dom[0],x_dom[1],100)[:,None]
    yy = np.ones_like(xx)*y
    pi_cond = pi.joint_pdf(np.hstack((yy,xx)))
    pi_cond = pi_cond/np.trapz(pi_cond, dx=(xx[1]-xx[0])) #, x=xx)
    
    if load_data == True:
        data = np.load('banana_data_yloc'+str(y)+'.npy')
        YX_transp_icnn = data[0]
        YX_transp_nn   = data[1]
        YX_transp_ott  = data[2]
    else:
        YX_transp_icnn = ot_icnn.evaluate(data_cond_y)
        YX_transp_nn = ot_nn.evaluate(data_cond_y)
        YX_transp_ott = ot_ott.evaluate(data_cond_y)
        np.save('banana_data_yloc'+str(y),[YX_transp_icnn, YX_transp_nn, YX_transp_ott])
        
    ax[2,i].hist(YX_transp_icnn[:,1], bins=nbins, density=True, label='$MLP$',alpha=.75, color='#1f77b4')
    ax[2,i].plot(xx, pi_cond, c='black', linewidth=lw)
    ax[2,i].set_ylim(0, 1.0)
    ax[2,i].set_xlim(x_dom[0],x_dom[1])
    if i==0:
        ax[2,i].tick_params(bottom =True, labelbottom = True, left=True, right=False, labelleft = True, labelsize=14)        
    else:
        ax[2,i].tick_params(left = False, right = False, bottom=True, labelbottom=True, labelleft=False)
    
    ax[1,i].hist(YX_transp_nn[:,1], bins=nbins, density=True, label='$NN$',alpha=.75,color='#ff7f0e')
    ax[1,i].plot(xx, pi_cond, c='black', linewidth=lw)
    ax[1,i].set_ylim(0, 1.0)
    ax[1,i].set_xlim(x_dom[0],x_dom[1])
    if i==0:
        ax[1,i].tick_params(bottom =False, labelbottom=False, left=True, right=False, labelleft = True, labelsize=14)        
    else:
        ax[1,i].tick_params(left = False, right = False , labelleft = False, bottom=False, labelbottom = False) #labelbottom = False, bottom = False) 
    
    ax[0,i].hist(YX_transp_ott[:,1], bins=nbins, density=True, label='$EOT$',alpha=.75,color='#d62728')
    ax[0,i].plot(xx, pi_cond, c='black', linewidth=lw)
    ax[0,i].set_ylim(0, 1.0)
    ax[0,i].set_xlim(x_dom[0],x_dom[1])
    if i==0:
        ax[0,i].tick_params(bottom =False, labelbottom=False, left=True, right=False, labelsize=14)        
    else:
        ax[0,i].tick_params(left = False, labelleft=False, right = False, bottom=False, labelbottom = False, labelsize=14) 
    
    
# Creating a single legend for all plots
handles, labels = [], []
for a in ax[:,0]:  # Flatten to iterate over all subplots
    for h, l in zip(*a.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)

print(handles)
print(labels)

# Adding a single legend outside the subplots
fig.legend(handles, labels, loc='upper right', bbox_to_anchor=(0.53, 0.94), fontsize=12)

# Adding row titles
ax[0,0].set_title('$x_1=-0.5$', fontsize=18)
ax[0,1].set_title('$x_1=3$', fontsize=18)

plt.tight_layout()
plt.show() #plt.savefig('conditionals_banana_vertical.pdf')
```
